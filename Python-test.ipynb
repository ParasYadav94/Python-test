# Step 1:- importing the libraries 

```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib as plt

## to import the data from the given data source using NSEpy package
import nsepy as ns
from datetime import date, datetime
```

# Step 2:- Importing the data from the given data source

```python
## A) Importing the data for NSE  stocks (INFY,TCS) between 2014-2016. Data level - Daily.
infy ,tcs= (ns.get_history(symbol = s, start= date(2015,1,1),
                          end= date(2016,12,31))for s in ["INFY","TCS"])
```

```python
infy # to view the data frame
# it has 495 rows * 14 columns 
```

```python
tcs # to view the TCS data
# it has 495 rows * 14 columns
```

```python
## B) IMPORTING OCLHV DATA FOR "NIFTY IT" INDEX, DATA LEVEL- DAILY

nifty= ns.get_history(symbol= "NIFTY IT",start= date(2015,1,1),
                            end= date(2016,12,31),index= True)
```

```python
nifty ## to view the imported data  
# it has 495 rows * 6 columns 
```

# The required data for all symbols have been imported 

# PART 1:

```python
# 1. Create 4,16,....,52 week moving average (closing price) for each stock and index. Through a defined function

## Creating a dataframe consisting of the Closing Price of all stocks/index
data= pd.DataFrame({"INFY": infy['Close'],
                  "TCS": tcs['Close'],
                   "NIFTY_IT": nifty['Close']})
data.tail(5)
```

```python
## to plot all three together
import matplotlib.pyplot as mlt
mlt.rcParams['figure.figsize']=[15,10]
data.plot(secondary_y= ["INFY","NIFTY_IT"],grid=True)
```

```python
## creating moving averages from mean with rolling window= 10
rolling_10= np.round(data.rolling(window=10,center= False).mean(),2)
rolling_10
```

```python
## Creating moving averages from mean with rolling window= 75
rolling_75= np.round(data.rolling(window=75, center= False).mean(),2)
rolling_75
```

# PART-2 :- DATA VISUALIZATION

```python
import bokeh as bk
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show, output_file
```

```python
## PART -3 

df= pd.DataFrame(data["INFY"])
df= df.fillna(0)
df.index= pd.to_datetime(df.index)
df.index.name= 'Date'
df.sort_index(inplace=True)
source = ColumnDataSource(df)
p= figure(x_axis_type="datetime", plot_width=500, plot_height=350)
p.line('Date','INFY',source= source)
output_file("ts.html")
show(p)
```
